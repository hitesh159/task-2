```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
weather_df = pd.read_csv('testset.csv')
```

```python
weather_df.head()
```

```python
weather_df['datetime_utc'] = pd.to_datetime(weather_df['datetime_utc'])
weather_df.set_index('datetime_utc', inplace=True)
```

```python
weather_df = weather_df.resample('D').mean()
```

```python
#Detecting Outliers
weather_df[weather_df.apply(lambda x: np.abs(x-x.mean()) / x.std() < 3).all(axis=1)]
```

```python
weather_df.count()
```

```python
weather_df.plot(y=' _fog')
```

```python
weather_df.groupby(' _tornado').count()
```

```python
to_drop = [' _snow', ' _tornado', ' _precipm']
weather_df.drop(to_drop, inplace=True, axis=1)
```
